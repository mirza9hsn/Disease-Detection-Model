#PreProcessing


```python
import pandas as pd

# Load dataset
symptoms_df = pd.read_csv('Datasets/symtoms_df.csv')

# Extract all unique symptoms from Symptom_1 to Symptom_4
symptom_columns = set()
for column in ['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4']:
    symptom_columns.update(symptoms_df[column].dropna().str.strip().unique())
symptom_columns = sorted(symptom_columns)  # Sort symptoms for consistency

# Initialize a new DataFrame with all symptom columns set to 0
processed_df = pd.DataFrame(0, index=symptoms_df.index, columns=symptom_columns)

# Iterate through the symptoms in the original dataset and set corresponding columns to 1
for index, row in symptoms_df.iterrows():
    for symptom in ['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4']:
        if pd.notna(row[symptom]):  # Check if the symptom is not NaN
            symptom_name = row[symptom].strip()
            if symptom_name in processed_df.columns:
                processed_df.loc[index, symptom_name] = 1

# Add the "prognosis" column based on the "Disease" column in symptoms_df
processed_df['prognosis'] = symptoms_df['Disease']

# Save the processed DataFrame to a new CSV file
output_file_path = 'Datasets/Processed_Symptoms.csv'
processed_df.to_csv(output_file_path, index=False)

print(f"Processed dataset saved to {output_file_path}")
```

#Train Test Split

```python
df = pd.read_csv("Datasets/Processed_Symptoms.csv")
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder


X = df.drop('prognosis', axis=1)
y = df['prognosis']

# ecoding prognonsis
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
    
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=20)
```

#Training Models 

```python

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np


# Create a dictionary to store models
models = {
    'SVC': SVC(kernel='linear'),
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'GradientBoosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'KNeighbors': KNeighborsClassifier(n_neighbors=5),
    'MultinomialNB': MultinomialNB()
}

# Loop through the models, train, test, and print results
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Test the model
    predictions = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, predictions)
    print(f"{model_name} Accuracy: {accuracy}")

    # Calculate confusion matrix
    cm = confusion_matrix(y_test, predictions)
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(cm, separator=', '))

    print("\n" + "="*40 + "\n")
```

```python
svc_model = SVC(kernel='linear')  # You can change kernel to 'rbf', 'poly', etc. if needed
svc_model.fit(X_train, y_train)
ypred = svc_model.predict(X_test)
accuracy_score(y_test, ypred)
```

```python
# Save the trained model to a file
import joblib

model_save_path = "Model/Disease_Prediction_Model.pkl"  # Replace with desired path
joblib.dump(svc_model, model_save_path)

print(f"Model saved to {model_save_path}")
```
